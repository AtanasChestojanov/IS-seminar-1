### Initial solution generator functions

#### We have 4 different functions that generate initial solution assignment matrices:
* "solution_capacity_min_reviews" generates solutions that try satisfy the maximum review capacity per person and minimum number of reviews per paper constraints.

* "solution_randomize_general" generates solutions in which we have controlled randomness, meaning we randomly shuffle reviewers and papers to create different staring points and we randomly skip some assingments or change tie-breaking criteria.

* "solution_general" generates a single solution per dataset. First it is iterating through the papers. Then it finds potential reviewers for the paper based on the preference matrix. Then for each of the reviewers we check the other constraints. If the cosntraints are met, we are assigning the paper to the reviewer. If the minimum reviewers for paper constraint is met, we go to the next paper.

* "solution_random" randomly assigns papers to a reviewer withour exceeding the reviewer capacity. 

* "solution_authorship_constraint" genereates a solution that tries to satisfy the authorship constraint.

* "solution_frinedship_constraint" genereates a solution that tries to satisfy the friendship constraint.

* "solution_reviewers_constraint" genereates a solution that tries to satisfy the minimum and maximum reviewers per paper constraint.

* "solution_combined_constraint" combines the last 3 functions.

### Crossover and mutation functions

* #### Crossover
        Crossover functions in genetic algorithms combine genetic information from two parents to create offspring, enhancing diversity. They mix genes from parents to explore the solution space. Examples include uniform crossover, where each gene is randomly selected from either parent; single-point crossover, which splits genes at one point; and two-point crossover, which uses two split points. PyGAD provides built-in options like single-point, two-point, uniform, and scattered crossover, and also supports custom functions for tailored behavior.

        For the purpose of this assignment we have implemented some more customized versions of the following crossover functions: column-wise crossover, row-wise crossover, uniform crossover, heuristic crossover, matrix crossover and two-point crossover.
        - Column-wise Crossover: Offspring are created by selecting entire columns alternately from two parent matrices.
        - Row-wise Crossover: Offspring are created by selecting entire rows alternately from two parent matrices.
        - Uniform Crossover: Each gene in the offspring is randomly selected from either parent, with equal probability.
        - Heuristic Crossover: Combines parents based on their fitness, favoring genes from the better-performing parent.
        - Matrix Crossover: Genes are mixed between parents by dividing the matrices into submatrices and swapping blocks.
        - Two-point Crossover: Genes are split at two points, and the middle segment is swapped between parents to form offspring.

* #### Mutation
        Mutation functions in genetic algorithms introduce randomness by altering genes in offspring, promoting diversity and avoiding premature convergence. They modify parts of a solution to explore the search space further. Examples include uniform mutation, where individual genes are flipped randomly; row-wise mutation, which alters entire rows in a matrix; and column-wise mutation, which alters entire columns. PyGAD supports built-in mutation methods and allows custom mutation functions for specific problem requirements.

        For the purpose of this assignment we have implemented some more customized versions of the following mutation functions: authorship-based mutation, friendship-based mutation, row-wise mutation, column-wise mutation, block mutation and swap mutation.
        - Authorship-based Mutation: Ensures no reviewer is assigned to papers they authored by flipping such assignments to 0.
        - Friendship-based Mutation: Removes assignments where friends are reviewing the same paper by altering conflicting genes.
        - Row-wise Mutation: Alters random rows in the assignment matrix, changing reviewer-paper allocations for selected rows.
        - Column-wise Mutation: Alters random columns in the assignment matrix, modifying reviewer assignments for specific papers.
        - Block Mutation: Changes a submatrix block of the assignment matrix by flipping its values.
        - Swap Mutation: Select two random indices and swaps their values.
