## Optimizing Conference Paper

### Problem definition

#### The goal is to optimize the assignment of reviewers to conference papers such that the assignements respect constrains, such as:

* Reviewer Capacity: Each reviewer can review a maximum of "capacity" papers.

* Minimum and Maximum Reviews per Paper: Each paper must have at least "min_reviews" and at most "max_reviews" reviewers.

* Avoid conflicts: Reviewers cannot review their own papers, they cannot review papers that are authored by their friends and friends cannot review the same paper.

#### The problem of assigning reviewers to conference papers is a combinatorial optimization problem with complex constraints. Finding an optimal solution manually or using brute-force methods is computationally expensive, especially as the number of reviewers and papers grows. GAs are well-suited to this problem because of its flexibillity, scalability, heuristic nature and fitness-based optimization.

While constructing this algorithm, each individual in the population represents a possible assignment of reviewers per papers, encoded as a binary matrix. GAs are applied as follows:

* Fitness Function: Evaluates assignments based on the given constrains of the given problem.

* Selection: Prioritizes assignments with better fitness scores for reproduction.

* Crossover: Combines parts of two assignments to create new, potentially better assignments.

* Mutation: Modifies assignments to explore solutions outside the current population and prevent premature convergence.

### Solution representation

#### The solution is represented as a 2D matrix in which rows represent reviewers and the columns represent papers. If the solution is a binary matrix S, then the cell S[i,j] tells us if the reviewer 'i' is assigned to the paper 'j' (1), or not (0).

### Initial solution generator functions

#### We have 4 different functions that generate initial solution assignment matrices:

* "solution_capacity_min_reviews" generates solutions that try satisfy the maximum review capacity per person and minimum number of reviews per paper constraints.

* "solution_randomize_general" generates solutions in which we have controlled randomness, meaning we randomly shuffle reviewers and papers to create different staring points and we randomly skip some assingments or change tie-breaking criteria.

* "solution_general" generates a single solution per dataset. First it is iterating through the papers. Then it finds potential reviewers for the paper based on the preference matrix. Then for each of the reviewers we check the other constraints. If the cosntraints are met, we are assigning the paper to the reviewer. If the minimum reviewers for paper constraint is met, we go to the next paper.

* "solution_random" randomly assigns papers to a reviewer withour exceeding the reviewer capacity. 

* "solution_authorship_constraint" genereates a solution that tries to satisfy the authorship constraint.

* "solution_frinedship_constraint" genereates a solution that tries to satisfy the friendship constraint.

* "solution_reviewers_constraint" genereates a solution that tries to satisfy the minimum and maximum reviewers per paper constraint.

* "solution_combined_constraint" combines the last 3 functions.

### Parent selection type

#### Parent selection is a step in Genetic Algorithms (GAs), where individuals (solutions) are chosen from the population to produce offspring for the next generation. The goal is to ensure that better solutions have a higher chance of contributing to the next generation while maintaining diversity to explore the search space effectively.

Types of Parent Selection Methods:

* Roulette Wheel Selection: Each individual's probability of being selected is proportional to its fitness score.

* Tournament Selection: Randomly select a subset (tournament) of individuals from the population and the individual with the highest fitness in the subset is chosen as a parent.

* Rank-Based Selection: Rank individuals by fitness and assign probabilities based on their rank, not their raw fitness value.

* Stochastic Universal Sampling (SUS): Similar to roulette wheel selection but ensures a more even distribution of selection pressure.

* Elitism with Selection: Guarantees that the top-performing individuals automatically survive to the next generation.

* Random Selection: Parents are chosen entirely at random from the population.

* Truncation Selection: Selects the top-performing individuals from the population, based on a pre-defined cutoff threshold.
